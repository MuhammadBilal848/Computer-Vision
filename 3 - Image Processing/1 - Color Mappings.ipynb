```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# in the 1970's HSL(Hue,Satuartion,Light) and HSV(Hue,Saturation,Value) were developed as alternative color models.
# HSL and HSV are more closely alligned with the way human vision percieves color.

# HSL Cylinder Model
# For better understanding see Image here 
# [ https://www.google.com/imgres?imgurl=https%3A%2F%2Frgbtohex.page%2Fimgs%2Fhsl-cylinder.png&imgrefurl=https%3A%2F%2Frgbtohex.page%2Fhsl&tbnid=_vUnfUzc9vpirM&vet=12ahUKEwifg5DYoNrtAhWYwoUKHfILBK4QMygUegUIARDQAQ..i&docid=ZGLFpDF_WbPICM&w=500&h=400&q=HSL&ved=2ahUKEwifg5DYoNrtAhWYwoUKHfILBK4QMygUegUIARDQAQ ]
# the way HSL model works is we have hue saturation and lightness and an HSV we're going to have hue saturation and then value.
# So hue saturation. Basically what this is is all around the cylinder.
# There's going to be the actual different color that is the green the yellow the red what actual hue
# we're looking for and then saturation is just how much of that hue you end up having and lightness is
# how dark or light it is. So on the top of the cylinder all the values are actually pure white.
# They're all just as light. So the absolute lightness you can get is just pure white and at the bottom of the cylinder we
# have pure black which is essentially no lightness in the color regardless of its saturation or hue.

# HSV Cylinder Model
# For better understanding see Image here 
# https://www.google.com/imgres?imgurl=http%3A%2F%2F1.bp.blogspot.com%2F-uKkPvILBxUc%2FVXThTiWG8LI%2FAAAAAAAACHg%2FIMJtFdkVPRU%2Fs1600%2Fimages.jpg&imgrefurl=http%3A%2F%2Fmeeporoi.blogspot.com%2F2015%2F06%2Fwhat-is-saturation.html&tbnid=bxnA83JZ3i0CnM&vet=12ahUKEwifg5DYoNrtAhWYwoUKHfILBK4QMygPegUIARDGAQ..i&docid=9JDG7chxqr5RsM&w=259&h=194&q=HSL&ved=2ahUKEwifg5DYoNrtAhWYwoUKHfILBK4QMygPegUIARDGAQ
# This is similar to an HSV cylinder model except instead of having a lightness we have a value.
# So again we have the hue color all around the circle and then saturation is going to define how close
# we are to white versus how Saturated we are in a particular hue.
# So the top is now basically the same hue all around and we're just deciding on a particular value
# depending on the hue and saturation.
# And you may have played of some color pictures before that followed the sort of model but notice now
# the White is actually at the very center of that cylinder.
# So essentially no saturation of any hue by default is going to be white.
```
