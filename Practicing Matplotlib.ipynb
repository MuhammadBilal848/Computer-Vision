### Functional Method for plotting

```python
np.random.seed(0)
x = np.random.randint(0,10,10)
y = x*2
```

```python
plt.plot(x,y)
# plt.show()  # use this with other ide's it will work here as well but we dont need here because jupter is showing its own.
```

```python
plt.plot(x,y,'r')
# **Colors**

# The following color abbreviations are supported:
# =============    ===============================
# character        color
# =============    ===============================
# ``'b'``          blue
# ``'g'``          green
# ``'r'``          red
# ``'c'``          cyan
# ``'m'``          magenta
# ``'y'``          yellow
# ``'k'``          black
# ``'w'``          white
# =============    ===============================
# If the color is the only part of the format string, you can
# additionally use any  `matplotlib.colors` spec, e.g. full names
# (``'green'``) or hex strings (``'#008000'``).
```

```python
plt.plot(x,y,color='cyan')
```

```python
# we can also set labels for x and y axis and title for graph as well
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.title('Acceleration')
```

```python
# what if we want to create multiple plots on the same plane, for this we use subplot function of matplotlib
# there are three main arguments in subplot row , column , index
# row*column will decide no of plots available on plane
# by using index number we can plot on each section
# E.g. subplot(2,2,1)
# 2*2 = 4 which means we have 4 sections to plot 
# we can use index number to plot on every section [keep in mind index starts at 1] 
# subplot(x,y,1)
# subplot(x,y,2)
# subplot(x,y,3)
# subplot(x,y,4)
x = np.linspace(0,10,10)
y = np.linspace(10,20,10)
plt.subplot(2,2,1)
plt.plot(x,y,'blue')
plt.subplot(2,2,2)
plt.plot(x,np.sqrt(y),'cyan')
plt.subplot(2,2,3)
plt.plot(x*np.pi,np.sqrt(y),'magenta')
plt.subplot(2,2,4)
plt.plot(np.sqrt(x*np.pi),np.sqrt(y),'yellow')
# plt.subplot(2,2,5) # this will generate error because the mutiple of row*col is 4 not 5
# plt.plot(np.sqrt(x*np.pi),np.sqrt(y),'yellow')
```

### Object-Oriented Method for plotting

```python
fig = plt.figure() # this means we have created a blank canvas for plotting
# # now we are adding list of axes to the plot [left, bottom,width,height]
# # all the values should be between 0 to 1(that doesn't mean that we cant use any other number we can use any number from 0
# # to n but the thing is that the graph will become long and we need to scroll it and thats not a good practice)
# # Be in between 0 and 1 is a good practice
axes = fig.add_axes([0.2, 0.2, 0.9, 0.9])
# # In the above code, axes([0.2, 0.2, 0.9, 0.9]), where the first ‘0.2’ indicates the distance between the left side axis 
# # and the border of the figure window/ Blank canvas that is 20%, of the total width of the figure window/ Blank canvas.
# # The second ‘0.2’ is indicating the distance between the bottom side axis and the border of the figure window that is 20%,
# # of the total height of the figure window.
# # Also, The first ‘0.9’ indicates that the axes width from left to right is 90% and
# # the latter ‘0.9’ indicates the axes height from the bottom to the top is 80%.
# axes2 = fig.add_axes([0.3,0.2,0.5,0.5])
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# we can plot graph on that just like we did before
axes.plot(np.sqrt(x),y)
# we can also set labels and titles
axes.set_xlabel('sqrt of x')
axes.set_ylabel('y')
axes.set_title('First')

# here we can also add more axes
axes2 = fig.add_axes([0.2,0.2,0.6,0.6])
axes2.plot(x,np.sqrt(y))
axes2.set_xlabel('x')
axes2.set_ylabel('sqrt of y')
axes2.set_title('Second')
```

```python
# we can use figsize parameter of figure to control the width and height of our plot
# plt.figure(figsize = (width,height))
# Width – Here, we have to input the width in inches. The default width is 6.4 .To broaden the plot, set the width
# greater than 1. And to make the graph less broad, set the width less than 6. 
# Height – Here, we have to input the height of the graph. The default value is 4.8 .To increase the length, set the 
# height greater than 4, and to decrease the height set the height less than 4. 
fig = plt.figure(figsize=(6,3))
# here now the width is 8 inches long and height is 3 inches long
fig.add_axes([0,0,1,1])

# we can make it wider
fig2 = plt.figure(figsize=(8,3))
fig2.add_axes([0,0,1,1])
# not he width is 8 inches long and height is 3 inches long

# lets try with different heightst
fig3 = plt.figure(figsize=(3,6))
fig3.add_axes([0,0,1,1])
# not he width is 3 inches long and height is 8 inches long

fig4 = plt.figure(figsize=(3,3))
fig4.add_axes([0,0,1,1])
# not he width is 3 inches long and height is 3 inches long
 
# more to read here [ https://www.pythonpool.com/matplotlib-figsize/ ]
# or watch this YT video [ https://www.youtube.com/watch?v=UUy6_ElQXBY&ab_channel=KimberlyFessel ]
```

```python
# we can also set dpi(dots per inch or pixel per inch).
# Dots per inches (dpi) determines how many pixels the figure comprises. The default dpi in matplotlib is 100.
fig = plt.figure(figsize=(5,5),dpi=150)
axes = fig.add_axes([0,0,1,1])
axes.plot(y,x)
```

```python
# we can also update all of our future figures , by using rcParams
# plt.rcParams['The parameter you want to update'] = (values,values,values,...)
plt.rcParams['figure.figsize'] = (5,5) # this will set all my future figures to 5 inch width and 5 inch height
plt.plot(x,y)

# you can read more about dpi here ,
# [ https://stackoverflow.com/questions/47633546/relationship-between-dpi-and-figure-size#:~:text=Dots%20per%20inches%20(dpi)%20determines,dpi%20in%20matplotlib%20is%20100. ] 
```

```python
plt.plot(x,np.sqrt(y)) # figsize is 5 inch width and 5 inch height
```

```python
# a function plt.subplots() helps to create subplots and add axes automatically to it, it returns fig and axes
# fig is again an empty canvas and axes are used for subplots
# so its more handy to use it rather than using plt.figure()
# it takes two main argument nrows and ncols
# the function will adjust axes on its own
# fig,axes = plt.subplots(2,2)
# the way subplots works is that 'fig,axes = plt.subplots(2,2)' this will create a blank canvas and draw 2*2= 4 subplots
# and save those plots in axes in form of object, so basically axes is a list of objects that are plots
# we can unpack axes to use them for plotting by using this syntax - fig,((ax1,ax2),(ax3,ax4))
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2)
ax1.plot(x,np.pi*y)
ax2.plot(np.sqrt(x),np.pi*y)
ax3.plot(x**4,np.pi*y)
ax4.plot(x**8,y)
```

```python
# there is another way to plot individually ,
fig,axes = plt.subplots(2,2)
print(axes) # list of objects that are plots 
axes[0,0].plot(x,np.pi*y,color='orange')
axes[0,0].set_title('First')
axes[0,1].plot(np.sqrt(x),np.pi*y,color='Magenta')
axes[0,1].set_title('Second')
axes[1,0].plot(x**4,np.pi*y,color='Cyan')
axes[1,0].set_title('Third')
axes[1,1].plot(x**8,y,color='Pink')
axes[1,1].set_title('Fourth')

plt.tight_layout() # this will create a gap between subplots plots 
# pros recommend to always put this line at the end your code , when you are dealing with subbplots

# You can read more about it on [ https://stackoverflow.com/questions/31726643/how-do-i-get-multiple-subplots-in-matplotlib ]
```

```python
# we can also set figsize for subplots as well
fig_four,axes = plt.subplots(2,2,figsize = (12,6))
axes[0,0].plot(x,np.pi*y,color='orange')
axes[0,0].set_title('First')
axes[0,1].plot(np.sqrt(x),np.pi*y,color='Magenta')
axes[0,1].set_title('Second')
axes[1,0].plot(x**4,np.pi*y,color='Cyan')
axes[1,0].set_title('Third')
axes[1,1].plot(x**8,y,color='Pink')
axes[1,1].set_title('Fourth')
```
